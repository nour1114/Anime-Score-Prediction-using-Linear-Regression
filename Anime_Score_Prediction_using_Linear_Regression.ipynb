```python
# Import required libraries
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Sample dataset
# You can replace this with:
# anime_data_extracted = pd.read_csv("your_file.csv")

data = {
    "Episodes": [12, 24, 36, 50, 75, 100],
    "Score": [7.1, 7.5, 7.8, 8.0, 8.4, 8.7]
}

anime_data_extracted = pd.DataFrame(data)

# Function
def homework(anime_data_extracted, X_column, Y_column):
    model = linear_model.LinearRegression()

    # Features and target
    X = anime_data_extracted[[X_column]]
    y = anime_data_extracted[Y_column]

    # Train model
    model.fit(X, y)

    # Prediction
    predictions = model.predict(X)

    # Accuracy Score
    result = model.score(X, y)

    # Extra evaluation metrics
    mse = mean_squared_error(y, predictions)
    r2 = r2_score(y, predictions)

    print("Model Trained Successfully ✅")
    print(f"Coefficient: {model.coef_[0]}")
    print(f"Intercept: {model.intercept_}")
    print(f"Mean Squared Error: {mse}")
    print(f"R² Score: {r2}")

    return float(result)

# Run the function
accuracy = homework(anime_data_extracted, "Episodes", "Score")

print(f"\nFinal Accuracy Score: {accuracy}")
```
